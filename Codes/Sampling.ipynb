```python
from google.colab import files

uploaded = files.upload()

for filename in uploaded.keys():
    print(f'Uploaded file: "{filename}" with {len(uploaded[filename])} bytes')

import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

# Specify the sampling factor (e.g., 2 for every 2nd pixel)
sampling_factor = 1

# Load the uploaded image
image_path = list(uploaded.keys())[0]
original_image = Image.open(image_path)

# Convert the image to a NumPy array
image_array = np.array(original_image)

# Display the original image
plt.figure(figsize=(8, 8))
plt.subplot(121)
plt.imshow(image_array)
plt.title("Original Image")

# Sample the image by selecting every n-th pixel
sampled_image = image_array[::sampling_factor, ::sampling_factor, :]

# Display the sampled image
plt.subplot(122)
plt.imshow(sampled_image)
plt.title(f"Sampled Image (Sampling Factor: {sampling_factor})")

plt.show()
```

```python
from google.colab import files

uploaded = files.upload()

for filename in uploaded.keys():
    print(f'Uploaded file: "{filename}" with {len(uploaded[filename])} bytes')

import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

#Sampling factor Initialize (Input)
sampling_factor = 16

#Image ka path declare kiya
image_path = list(uploaded.keys())[0]

#Image ke path se image le liya
original_image = Image.open(image_path)

#image ko array mein convert kiya
image_array=np.array(original_image)

#original image aur sampled image ko ek sath rakhne ke liye jagah banaya
plt.figure(figsize=(8,8))

#Divide karke pehla konsa ye bolna hai
plt.subplot(121)
plt.imshow(original_image)
plt.title("original image")

#samlping kar diya by nth wala method
sampled_image=image_array[::sampling_factor,::sampling_factor,:]

#Second image konsa ye bolna hai
plt.subplot(122)
plt.imshow(sampled_image)
plt.title("Sampled Image (sampling factor = {sampling_factor})")

#Showing both images
plt.show()
```
