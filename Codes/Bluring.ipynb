*****Img Blur function of open CV*****

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from google.colab.patches import cv2_imshow

from google.colab import files
uploaded = files.upload()

for filename in uploaded.keys():
    print(f'Uploaded file: "{filename}" with {len(uploaded[filename])} bytes')

#Creating a kernel
kernel1=np.array(([0,0,0],[0,1,0],[0,0,0]))

#Convuling image
image=cv2.imread(filename)

img_blur=cv2.filter2D(src=image,ddepth=-1,kernel=kernel1)

cv2_imshow(image)
cv2_imshow(img_blur)

```

*****Gaussian Bluring*****

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from google.colab.patches import cv2_imshow

from google.colab import files
uploaded = files.upload()

for filename in uploaded.keys():
    print(f'Uploaded file: "{filename}" with {len(uploaded[filename])} bytes')

#read image
image=cv2.imread(filename)

#gaussian bluring
gaussian_blur=cv2.GaussianBlur(src=image,ksize=(5,5),sigmaX=0,sigmaY=0)

#Showing image
cv2_imshow(image)
cv2_imshow(gaussian_blur)
```

*****Bilateral Filter*****

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from google.colab.patches import cv2_imshow

from google.colab import files
uploaded = files.upload()

for filename in uploaded.keys():
    print(f'Uploaded file: "{filename}" with {len(uploaded[filename])} bytes')

#read image
image=cv2.imread(filename)

#Bilateral filtering
bilateral_Filter=cv2.bilateralFilter(src=image,d=12,sigmaColor=0,sigmaSpace=300)

#Showing
cv2_imshow(image)
cv2_imshow(bilateral_Filter)
```
