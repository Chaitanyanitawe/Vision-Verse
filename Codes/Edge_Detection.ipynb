*****Laplacian Edge Detection*****

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from google.colab.patches import cv2_imshow

from google.colab import files
uploaded = files.upload()

for filename in uploaded.keys():
    print(f'Uploaded file: "{filename}" with {len(uploaded[filename])} bytes')

#Read Image
image1=cv2.imread(filename,cv2.IMREAD_COLOR)

#Apply Gaussian Blur
image1_blur=cv2.GaussianBlur(src=image1,ksize=(3,3),sigmaX=0,sigmaY=0)

#Convert to gray(Not Compulsory)
image1_blur_gray=cv2.cvtColor(image1_blur,cv2.COLOR_BGR2GRAY)

#apply laplaian operation
image1_laplacian=cv2.Laplacian(src=image1_blur_gray, ddepth=cv2.CV_16S , ksize=3 )
abs_image1_laplacian = cv2.convertScaleAbs(image1_laplacian)

cv2_imshow(image1)

cv2_imshow(abs_image1_laplacian)
```

*****Hough Detection*****


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from google.colab.patches import cv2_imshow

from google.colab import files
uploaded = files.upload()

for filename in uploaded.keys():
    print(f'Uploaded file: "{filename}" with {len(uploaded[filename])} bytes')

#Read Image
image1=cv2.imread(filename,cv2.IMREAD_COLOR)

#Convert to gray
image_gray=cv2.cvtColor(image1,cv2.COLOR_BGR2GRAY)

#Gaussian Blur
image_blur=cv2.GaussianBlur(src=image_gray, ksize=(3,3),sigmaX=0,sigmaY=0)

#canny Edge
image_canny=cv2.Canny(image=image_blur,threshold1=5,threshold2=130)

#Hough Transform
img_hough=cv2.HoughLinesP(image_canny,1,np.pi/180, threshold=60,minLineLength=10,maxLineGap=250)

for line in img_hough:
   x1,y1,x2,y2= line[0]
   cv2.line(image1,(x1,y1),(x2,y2),(255,0,0),4)

#Show

cv2_imshow(img_hough)
cv2_imshow(image1)
```
