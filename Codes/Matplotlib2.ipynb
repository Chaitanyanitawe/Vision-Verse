```python
import matplotlib.pyplot as plt
import numpy as np
Cost = [33,12,55,67,88,99]
cars = [1,2,3,4,5,6]

cars_ticks=list(range(1,8))
cost_ticks=list(range(30,110,10))

plt.plot(cars,Cost)
plt.title("Cars Vs Price ", fontsize=24 )
plt.xlabel("Car No. -->")
plt.ylabel("Price (in USD) -->")
plt.xticks(cars_ticks , [f"Car{x}" for x in cars_ticks] )
plt.yticks(cost_ticks,[f"${x}K" for x in cost_ticks])

plt.show()
```

```python
Cost = [33,12,55,67,88,99]
cars = [1,2,3,4,5,6]
values = [16,18,58,45,25,12]
plt.figure(1)
plt.plot(Cost, label="Cost")
plt.legend( loc="best")

plt.figure(2)
plt.plot(cars, label="cars")
plt.legend( loc="best")

plt.figure(3)
plt.plot(values , label="values")

plt.legend( loc="best")

plt.show
```

```python
import matplotlib.pyplot as plt

# Data for the pie chart
labels = ['Category A', 'Category B', 'Category C']
sizes = [30, 45, 25]

# Specify colors for each category
colors = ['red', 'green', 'blue']

# Create a pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

# Show the pie chart
plt.show()




```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange((100))

fig, axs= plt.subplots(2,2)
axs[0,0].plot(x, np.sin(x))
axs[0,0].set_title("Sine")

axs[0,1].plot(x,np.log(x))

axs[1,0].plot(x, np.sin(x))

axs[1,1].plot(x, np.tan(x))

plt.show

plt.savefig("Plots.png",dpi=4,transparent=True)
```

```python
ax=plt.axes(projection="3d")


x = np.arange(0,50,0.1)
y = np.sin(x)
z = np.cos(x)

ax.plot(x,y,z)
ax.set_title("3D PLOT")
ax.set_xlabel("test")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a figure
fig = plt.figure()

# Add axes to the figure at [0,0,1,1]
ax = fig.add_axes([1, 0, 0.5, 1])

# Now, you can plot on the new axes (ax)
ax.plot([0, 1, 2, 3], [0, 1, 4, 9])

# Show the plot
plt.show()
```
