## Exercise 1

** Follow along with these steps: **
* ** Create a figure object called fig using plt.figure() **
* ** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* ** Plot (x,y) on that axes and set the labels and titles to match the plot below:**

```python
import matplotlib.pyplot as plt

x_values = [-0.04, -0.02, 0.00, 0.02, 0.04]
y_values = [2, 4, 6, 8, 10]

fig=plt.figure(1)

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x_values, y_values)

ax.set_xlabel('X-axis Label')
ax.set_ylabel('Y-axis Label')
ax.set_title('Plot Title')

plt.show()
```

## Exercise 2
** Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

** Now plot (x,y) on both axes. And call your figure object to show it.**

```python
import matplotlib.pyplot as plt

x_values = [-0.04, -0.02, 0.00, 0.02, 0.04]
y_values = [2, 4, 6, 8, 10]

fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])

ax1.plot(x_values, y_values, color='blue', label='ax1 Plot')

ax1.set_xlabel('X-axis Label')
ax1.set_ylabel('Y-axis Label')
ax1.set_title('ax1 Plot Title')
ax1.legend()

ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])

ax2.plot(x_values, y_values, color='red', label='ax2 Plot')

ax2.set_xlabel('X-axis Label')
ax2.set_ylabel('Y-axis Label')
ax2.set_title('ax2 Plot Title')
ax2.legend()

plt.show()
```

## Exercise 3

** Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python

import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])

x_values = np.linspace(0, 1, 10)
y_values = np.sin(x_values)
ax1.plot(x_values, y_values, label='Curve on ax1', color='blue')

ax1.set_xlabel('X-axis Label')
ax1.set_ylabel('Y-axis Label')
ax1.set_title('Plot on ax1')
ax1.legend()

ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])

x_values_sub = np.linspace(0, 1, 10)
y_values_sub = np.exp(x_values_sub)
ax2.plot(x_values_sub, y_values_sub, label='Curve on ax2', color='red')

ax2.set_xlabel('X-axis Label')
ax2.set_ylabel('Y-axis Label')
ax2.set_title('Plot on ax2')
ax2.legend()

plt.show()
```

## Exercise 4

** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
import matplotlib.pyplot as plt
import numpy as np

x_values = np.arange(0, 101, 20)

y_values_2x = 2 * x_values
y_values_x_squared = x_values ** 2

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

axes[0].plot(x_values, y_values_2x, label='y=2x', linestyle='dotted', color='blue')
axes[0].set_xlabel('X-axis Label')
axes[0].set_ylabel('Y-axis Label')
axes[0].set_title('y=2x')
axes[0].legend()

axes[1].plot(x_values, y_values_x_squared, label='y=x^2', linestyle='-', color='red')
axes[1].set_xlabel('X-axis Label')
axes[1].set_ylabel('Y-axis Label')
axes[1].set_title('y=x^2')
axes[1].legend()

plt.tight_layout()

plt.show()
```
